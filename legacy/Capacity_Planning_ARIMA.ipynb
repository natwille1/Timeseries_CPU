### ARIMA Model - CPU 

Joined Data Set

```python
# split into train and test sets
X = joined['avg_cpu'].values
train, test = X[1:100], X[len(X)-7:]
```

```python
# train autoregression
model = ARIMA(train, order=(10,1,0))
model_fit = model.fit()
print('Lag: %s' % model_fit.k_ar)
print('Coefficients: %s' % model_fit.params)
# make predictions
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
for i in range(len(predictions)):
    print('predicted=%f, expected=%f' % (predictions[i], test[i]))
error = metrics.mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot results
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```

### ARIMA Model - Mem

Joined Data Set

```python
joined.dropna(inplace=True)
#train_mem.dropna(inplace=True)
```

```python
X_mem = joined['avg_mem'].values
train_mem, test_mem = X_mem[1:len(X_mem)-7], X_mem[len(X_mem)-7:]
# train autoregression
model_mem = ARIMA(train_mem, order=(5,1,0))
model_fit = model_mem.fit()
print('Lag: %s' % model_fit.k_ar)
print('Coefficients: %s' % model_fit.params)
# make predictions
predictions = model_fit.predict(start=len(train_mem), end=len(train_mem)+len(test_mem)-1, dynamic=False)
for i in range(len(predictions)):
    print('predicted=%f, expected=%f' % (predictions[i], test_mem[i]))
error = metrics.mean_squared_error(test_mem, predictions)
print('Test MSE: %.3f' % error)
# plot results
pyplot.plot(test_mem)
pyplot.plot(predictions, color='red')
pyplot.show()
```
