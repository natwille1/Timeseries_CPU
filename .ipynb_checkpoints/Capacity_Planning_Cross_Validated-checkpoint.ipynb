### Data Analysis


```python
class Data(object):
    def __init__(self, data):
        self.data = data
    
    def convert_datetime(self, column):
        self.data[column] = pd.to_datetime(self.data[column])
        #return self.data
        
    def check_nans(self):
        nans = self.data[self.data.isnull().any(axis=1)]
        return nans
    
    def drop_nans(self, axis=0, inplace=True):
        self.data.dropna(axis=axis, inplace=inplace)
        
    def select_dates(self, start_time, end_time, time='_time'):
        selected = data.set_index(time).between_time(start_time = start_time, end_time = end_time)
        return selected
    
    def interpolate_data(self, inplace=True, time='_time', method='linear', order=2, limit_area='inside'):
        if inplace:
            if method=='polynomial' or method=='spline':
                self.data.set_index(time).interpolate(method=method, order=order, inplace=True)
            else:
                self.data.set_index(time).interpolate(method=method, limit_area=limit_area, inplace=True)
        else:
            if method=='polynomial' or method=='spline':
                interpolated = self.data.set_index(time).interpolate(method=method, order=order)
            else:
                interpolated = self.data.set_index(time).interpolate(method=method, limit_area=limit_area)
            return interpolated
        
    def adfuller_test(self, column):
        result = adfuller(self.data[column])
        print('ADF Statistic: %f' % result[0])
        print('p-value: %f' % result[1])
        print('Critical Values:')
        for key, value in result[4].items():
            print('\t%s: %.3f' % (key, value))
            
    def decompose_seasonal(self, column, model='additive', verbose=False, time='_time', freq=24):
        timedf = self.data.set_index(time)
        result = seasonal_decompose(timedf[column], model=model, freq=freq)
        if verbose:
            print(result.trend)
            print(result.seasonal)
            print(result.resid)
            print(result.observed)
        plt.figure(figsize=(20,8))
        result.plot()
        plt.show()
        return result
        
    def lag_plot(self, column):
        plt.figure(figsize=(8,6))
        lag_plot(self.data[column])
        plt.show()
        
    def plot_acf(self, column, lags=200):
        plt.figure(figsize=(8,6))
        plot_acf(self.data[column], ax=plt.gca(), lags=lags)
        plt.show()
        
    def plot_pacf(self, column, lags=200):
        plt.figure(figsize=(8,6))
        plot_pacf(self.data[column], ax=plt.gca(), lags=lags)
        plt.show()
        
    def autocorrelation_plot(self, column):
        plt.figure(figsize=(8,6))
        autocorrelation_plot(self.data[column])
        plt.show()
        
    def rolling_stats(self, column, window=12, plot=True, time='_time', return_data=True):
        timedf = self.data.set_index(time)[column]
        ma = self.data[[time, column]].rolling(on=time, window=window).mean()
        mstd = self.data[[time, column]].rolling(on=time, window=window).std()
        if plot:
            plt.figure(figsize=(8,6))
            orig = plt.plot(timedf,  label='Original')
            mean = plt.plot(ma.set_index(time),label='Rolling mean')
            mstd = plt.plot(mstd.set_index(time), color='black', label='Rolling std')
            plt.legend(loc='best')
            plt.title("Rolling mean and standard deviation")
            plt.show()
        if return_data:
            return ma, mstd
    
    def return_dataframe(self):
        return self.data


        
class BlockingTimeSeriesSplit():
    def __init__(self, n_splits): 
        self.n_splits = n_splits 

    def get_n_splits(self, X, y, groups): 
        return self.n_splits 

    def split(self, X, y=None, groups=None): 
        n_samples = len(X) 
        k_fold_size = n_samples // self.n_splits 
        indices = np.arange(n_samples) 
        margin = 0 
        for i in range(self.n_splits): 
            start = i * k_fold_size 
            stop = start + k_fold_size 
            mid = int(0.8 * (stop - start)) + start 
            yield indices[start: mid], indices[mid + margin: stop]      
        
    
```

### Stationarity

Next, let's check if the timeseries is stationary. 

```python
# Plot moving average and std 

selected.rolling_stats('cpu', window=24)
selected.rolling_stats('mem', window=24)

# Calculate "Augmented Dickey-Fuller" test on original data

selected.adfuller_test('cpu')
selected.adfuller_test('mem')

# Decompose time series 

cpu_decomposed = selected.decompose_seasonal('cpu', verbose=False)
mem_decomposed = selected.decompose_seasonal('mem')

# Calculate ADF test on detrended data (residuals)

cpu_decomposed.resid 
```

Calculating rolling mean and standard deviation reveals much stronger seasonality patterns than the original 1h interval. The ADF test also indicates that the timeseries are most likely not stationary, with a ADF statistic only slightly above the 1% critical value. We can qualitatively see from the rolling time plots as well. 

```python
# Calculate ACF and PACF for orginal data 

selected.plot_acf('cpu', lags=50)
selected.plot_acf('mem', lags=50)

selected.plot_pacf('cpu', lags=50)
selected.plot_pacf('mem', lags=50)

# Full autocorrelation plot for original data

selected.autocorrelation_plot('cpu')
selected.autocorrelation_plot('mem')

# Calculate ACD and PACF for decomposed data

# Residuals
plt.figure(figsize=(8,6))
plot_acf(cpu_decomposed.resid, ax=plt.gca(), lags=50)
plt.show()
plt.figure(figsize=(8,6))
plot_pacf(cpu_decomposed.resid, ax=plt.gca(), lags=50)
plt.show()

# Trend
plt.figure(figsize=(8,6))
plot_acf(cpu_decomposed.trend, ax=plt.gca(), lags=50)
plt.show()
plt.figure(figsize=(8,6))
plot_pacf(cpu_decomposed.resid, ax=plt.gca(), lags=50)
plt.show()

# Seasonal
plt.figure(figsize=(8,6))
plot_acf(cpu_decomposed.seasonal, ax=plt.gca(), lags=50)
plt.show()
plt.figure(figsize=(8,6))
plot_pacf(cpu_decomposed.resid, ax=plt.gca(), lags=50)
plt.show()

```

The ACF and PACF plots provide useful information about the type of models that are appropriate for these timeseries. The PACF function removes the intervening correlations between Xt and Xt+1, Xt+1 and Xt+2, etc. The partial autocorrelation at lag 2 is therefore the difference between the actual correlation at lag 2 and the expected correlation due to the propagation of correlation at lag 1. This allows us to remove the "chain" effect of correlations in timeseries, and identify autocorrelations between Xt and Xt at different lags more robustly. 

For the CPU dataset, the PACF shows a sharp decrease in correlation past lag 2, indicating that the higher order correlations in the ACF are most likely explained by lag 1/2 autocorrelation. 

For the Mem dataset, PACF also sharply decreases after lag 2, indicating similar autocorralation behaviour as the CPU timeseries. Interestingly, the ACF indicates stronger autocorrelations, with correlation coefficients > 0.4 until around the 24th lag. It also appears the correlation "pattern" repeats itself for lags between 24-48, indicating a strong daily seasonal component within this dataset. 

For the CPU-trend dataset, the ACF and PACF show no significant correlations. 

For the CPU-seasonal datatset, the ACF plot shows strong correlation patterns, repeated every 24 lags. Therefore, we can assume that the majority of the correlation seen in the original ACF plot is due to the seasonal component of the dataset. 

The plots suggest that AR models with around order 1-2 could be used for both timeseries. 

Looks like there are some interesting correlations across longer lag periods e.g ~180 for CPU. This corresponds to about every 25 days, which is close to a month. This could be a sign of longer term seasonality within the data, although the correlation strength seems to decay over time. That suggests the overall system is changing from its initial state (e.g adding more hardware, or minimising overall workload). We should take this into account during the model building process.

### Generating timeseries train-test datasets

We will experiment with two different approaches to generating the train and test datasets. The TimeSeries splitter class in sklearn returns indices of different lengths for each split iteration. Contrarily, the BlockTimeseries splitter returns indices of the same length for each split iteration, and tries to avoid test set leakage into the training dataset for subsequent splits. 

```python
## Timeseries splits

from sklearn.model_selection import TimeSeriesSplit

X = np.array(selected.return_dataframe())
y = selected.return_dataframe()

print("total days of data: {:.1f}".format(X.shape[0]/24))

## Keep last 20% for testing set 

size = int(len(X) * 0.8)
train = X[:size]
test = X[size:]

print("total days of training data: {:.1f}. Shape: {}".format(train.shape[0]/24, train.shape))
print("total dats of test data: {:.1f}. Shape: {}".format(test.shape[0]/24, test.shape))

split = 3
tscv = TimeSeriesSplit(n_splits = split)
btscv = BlockingTimeSeriesSplit(n_splits=split)

train_timeseries = []
test_timeseries = []

for train_index, test_index in tscv.split(train):
    print("TimeSeries split")
    X_train, X_test = train[train_index], train[test_index]
    X_train, X_test = pd.DataFrame(X_train), pd.DataFrame(X_test)
    X_train.columns = ['_time', 'CPU', 'Mem']
    X_test.columns = ['_time', 'CPU', 'Mem']
    train_timeseries.append([X_train])
    test_timeseries.append([X_test])
    ax1 = X_train.set_index('_time').plot()
    X_test.set_index('_time').plot(ax=ax1)
    #plt.plot(X_test)
    plt.show()
    print("X_train size: {}, X_test size: {}".format(X_train.shape, X_test.shape))
    
print()
print()
    
for train_index, test_index in btscv.split(train):
    print("Blocking split")
    X_train, X_test = train[train_index], train[test_index]
    X_train, X_test = pd.DataFrame(X_train), pd.DataFrame(X_test)
    X_train.columns = ['_time', 'CPU', 'Mem']
    X_test.columns = ['_time', 'CPU', 'Mem']
    ax1 = X_train.set_index('_time').plot()
    X_test.set_index('_time').plot(ax=ax1)
    #plt.plot(X_test)
    plt.show()
    print("X_train size: {}, X_test size: {}".format(X_train.shape, X_test.shape))
    
```

```python
# Generate an evaluation function 

def run_model(X, order):
    # train autoregression
    size = int(len(X) * 0.66)
    train, test = X[:size], X[size:len(X)]
    test = test[0:10]
   # print(test)
    predictions = []
    history = [x for x in train]
    for i in range(len(test)):
        model = ARIMA(history, order=order)
        model_fit = model.fit(disp=0)
#        residuals = pd.DataFrame(model_fit.resid)
        yhat = model_fit.forecast()[0]
        history.append(test[i])
        predictions.append(yhat)
    mse = metrics.mean_squared_error(test, predictions)
    return mse

def fast_model(X, order):
    size = int(len(X) * 0.66)
    train, test = X[:size], X[size:len(X)]
    train_time = train.drop('time', axis=1)
    test_time = test.drop('time', axis=1)
   # print(test)
    model = ARIMA(train_time, order=order)
    model_fit = model.fit()
    testset = test[0:51]
    if order[1] > 0:
        predictions = model_fit.predict(start=len(train_time), end=len(train_time)+50, typ='levels')
    else:
        predictions = model_fit.predict(start=len(train_time), end=len(train_time)+50)
    mse = metrics.mean_squared_error(testset['avg(Value)'], predictions)
    return mse 
    

def grid_search(data, p_values, d_values, q_values):
    best_score, best_cfg = float("inf"), None
    scores, orders = [],[]
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    mse = fast_model(data, order)
                    print(mse)
                    scores.append(mse)
                    orders.append(order)
                    if mse < best_score:
                        best_score, best_cfg = mse, order
                    print("ARIMA order: %s, SCORE %0.3f" % (order, mse))
                    #print("parameters: p = %s, q = %s, d = %s" % (p,q,d))
                except:
                    continue
    print("Best ARIMA order: %s, BEST SCORE: %0.3f" % (best_cfg, best_score))
    return scores, orders
        
```

### Use auto arima
In order to compare our grid search method for finding the optimal model with other methods, we can build other models using the he auto arima model implementation from the Pyramid web framework. 

As stated in the auto_arima docs:

The auto_arima function seeks to identify the most optimal parameters for an ARIMA model, and returns a fitted ARIMA model. This function is based on the commonly-used R function, forecast::auto.arima.

The auro_arima function works by conducting differencing tests (i.e., KwiatkowskiPhillipsSchmidtShin, Augmented Dickey-Fuller or PhillipsPerron) to determine the order of differencing, d, and then fitting models within ranges of defined start_p, max_p, start_q, max_q ranges. If the seasonal optional is enabled, auto_arima also seeks to identify the optimal P and Q hyper- parameters after conducting the Canova-Hansen to determine the optimal order of seasonal differencing, D.

```python
# automatically generate model with auto_arima library from Pyramid 
cpu_inter.dropna(inplace=True)


# Function expects a tuple for order, specifying start and end parameters for p and q 
def auto_model(order):
    size = int(len(cpu_inter) * 0.66)
    train, test = cpu_inter[:size], cpu_inter[size:len(cpu_inter)]
    train_time = train.drop('time', axis=1)
    test_time = test.drop('time', axis=1)
    p,q, end_p, end_q = order[0], order[1], order[2], order[3]
    print(p,q,end_p,end_q)
    stepwise_model = auto_arima(train_time, start_p=p, start_q=q,
                           max_p=end_p, max_q=end_q, trace=False,
                           error_action='warn',seasonal=False,
                           suppress_warnings=True, 
                           stepwise=True, max_order=None, maxiter=100)
    print("AIC: %f" % stepwise_model.aic())
    print("AICC: %f" % stepwise_model.aicc())
    print("BICC: %f" % stepwise_model.bic())
    return stepwise_model
#print("BSE: %f" % stepwise_model.bse())

```

```python
# generate automodel 
# Function expects a tuple for order, specifying start and end parameters for p and q 
order = (1,0,10,2)
automodel_d010 = auto_model(order)
order = (10,0,20,2)
automodel_d1020 = auto_model(order)
order = (20,0,24,2)
automodel_d2024 = auto_model(order)

```

```python
for i in range(3):
#     print("yes")
    aic = []
    aic.append(automodel_d010.aic())
    aic.append(automodel_d1020.aic())
    aic.append(automodel_d2024.aic())
    

print(aic)
```
